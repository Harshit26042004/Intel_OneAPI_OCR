```python
import keras_ocr
```

```python
pipeline=keras_ocr.pipeline.Pipeline()
```

```python
imgs=[
    keras_ocr.tools.read(imgs)for imgs in ["https://support.apple.com/library/content/dam/edam/applecare/images/en_US/iOS/ios15-iphone12-pro-messages-send-group-imessage.png",
                                        "https://ih1.redbubble.net/image.1940221577.0479/st,small,845x845-pad,1000x1000,f8f8f8.jpg"
                                        ]
]
```

```python
prediction=pipeline.recognize(imgs)
prediction
```

```python
import matplotlib.pyplot as plt
fig,axs=plt.subplots(nrows=len(imgs),figsize=(10,20))
for ax,images,predictions in zip(axs,imgs,prediction):
    keras_ocr.tools.drawAnnotations(image=images,predictions=predictions,ax=ax)
```
